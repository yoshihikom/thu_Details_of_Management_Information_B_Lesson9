```python
from pathlib import Path
import edinet

#データを取得して, PDF 版とXBRL 版を保存する
#ここでは、株式会社ブロードバンドタワーの有価証券報告書を取得する
#https://disclosure2.edinet-fsa.go.jp/ にアクセスして、ブロードバンドタワーで検索し、ドキュメント番号を調べます。
# 今回は、S100PXNE。
xbrl_path = edinet.api.document.get_xbrl("S100PXNE", save_dir=Path.cwd())
pdf_path = edinet.api.document.get_pdf("S100PXNE", save_dir=Path.cwd())
```

```python
from edinet.xbrl_file import XBRLFile
from edinet.parser.aspects.business import Business

#先程ダウンロードしたデータを参照する
xbrl = XBRLFile("./S100PXNE_1.xbrl")

#取得したxbrl データからparse する
content = xbrl.parse_by(Business).policy_environment_issue_etc
r_d = xbrl.parse_by(Business).research_and_development
```

```python
#経営方針、経営環境及び対処すべき課題等の情報を取り出す
print(content)
```

```python
#研究開発情報のデータを取り出す
print(r_d)
```
